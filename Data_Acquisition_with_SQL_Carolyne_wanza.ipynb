```python
import pandas as pd
```

```python
citiesdb = pd.read_csv('traffic_accidents.csv') 
citiesdb.shape
```

# New Section

```python
citiesdb = citiesdb[(citiesdb['VEHICLE_YEAR'] >= 2006)]
citiesdb.shape
```

```python
%sql DROP TABLE if EXISTS citiesdb;

%sql PERSIST citiesdb;

%sql SELECT * FROM citiesdb LIMIT 5;
```

```python
# Display a list of 5 makes that had the highest number of crashes?
%sql select make,count(*) no_of_crashes from citiesdb group by make order by no_of_crashes desc limit 5
```

```python
# Most personal vehicles involved in crashes were traveling towards which direction?
# VEHICLE_USE	TRAVEL_DIRECTION
%%sql select TRAVEL_DIRECTION,count(*) no_of_crashes from citiesdb where VEHICLE_USE='PERSONAL' 
group by travel_direction ORDER BY no_of_crashes desc

```

```python
#  How many passengers were involved in personal cars were in accidents?
%sql select sum(OCCUPANT_CNT) no_of_passengers from citiesdb where VEHICLE_USE='PERSONAL' 
```

```python
# How many passengers were involved in accidents while traveling in personal Sport Utility vehicles?
%sql select sum(OCCUPANT_CNT) no_of_passengers from citiesdb  where VEHICLE_TYPE='SPORT UTILITY VEHICLE (SUV)' 
```

```python
# Compute the total number of Mercedez-Benz vehicles involved in accidents while being parked.
%sql select COUNT(DISTINCT VEHICLE_ID) NO_OF_VEHICLES from citiesdb where MAKE='MERCEDES-BENZ' AND MANEUVER='PARKED'
```

```python
# Compute the total occupant count per vehicle count.
%sql select make,SUM(OCCUPANT_CNT) occupants_count from citiesdb group by make order by occupants_count desc
```

```python
# Sort the number of passengers in descending order by travel direction taking into account travel direction. Return travel direction and no. of passengers.
%%sql select travel_direction,sum(OCCUPANT_CNT) no_of_passengers from citiesdb 
group by travel_direction ORDER BY no_of_passengers desc
```

```python
# Select the top 5 vehicle models involved in the accident with no defect, order them by make, and show their maneuver
# %sql select * from citiesdb limit 2

%sql select make,model,maneuver,count(*) no_of_crashes from citiesdb where model in (select model from (select model,count(*) no_of_crashes from citiesdb where VEHICLE_DEFECT='NONE' and not model='UNKNOWN' and not model='OTHER (EXPLAIN IN NARRATIVE)' group by model order by no_of_crashes desc limit 5)) group by make,model,maneuver order by make

# %sql select model from (select model,count(*) no_of_crashes from citiesdb where VEHICLE_DEFECT='NONE' and not model='UNKNOWN' and not model='OTHER (EXPLAIN IN NARRATIVE)' group by model order by no_of_crashes desc limit 5)
```

```python
# Which ridesharing vehicles in an accident had the most number of passengers? Return model, and the number of passengers.
%sql select model,occupant_cnt from citiesdb where vehicle_use='RIDESHARE SERVICE' order by occupant_cnt desc limit 1
# %sql select * from citiesdb limit 1 
```
